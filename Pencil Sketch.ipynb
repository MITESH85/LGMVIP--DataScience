```python
import cv2
import matplotlib.pyplot as plt
```

### Import Image File

```python
image = cv2.imread('my.png')
```

```python
plt.imshow(image)
plt.axis(False)
plt.show()
```

```python
plt.imshow(image[:,:,::-1])
plt.axis(False)
plt.show()
```

### Converting the original image to greyscale:

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

### Invert the new grayscale image

```python
invert_image=cv2.bitwise_not(grey_image)
```

### Blur the image

```python
blur_image=cv2.GaussianBlur(invert_image, (111,111),0)
```

### Invert the blurred image and convert the image into a pencil sketch:

```python
invblur_image=cv2.bitwise_not(blur_image)
sketch_image=cv2.divide(grey_image,invblur_image, scale=256.0)
```

### Display both pencil sketch

```python
plt.imshow(sketch_image)
plt.axis(False)
plt.show()
```

### Save Sketch

```python
cv2.imwrite('sketch.png', sketch_image)
```
